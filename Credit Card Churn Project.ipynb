### Split Data Into Training and Testing

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm as svmc
from sklearn import metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
X = ccdf.iloc[:,1:]
y = ccdf.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.3, random_state = 23)
```

```python
X_train
```

```python
X_test
```

#### Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
y_preds_dt = dt.predict(X_test)


cmplt_dt = ConfusionMatrixDisplay(confusion_matrix(y_test, y_preds_dt))
cmplt_dt.plot()
print(classification_report(y_preds_dt, y_test))
```

#### Naive Bayes 

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred_gnb = gnb.predict(X_test)

cmplt_gnb = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_gnb))
cmplt_gnb.plot()
print(classification_report(y_pred_gnb, y_test))
```

#### K- Nearest Neighbors 

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_preds_knn = knn.predict(X_test)

cmplt_knn = ConfusionMatrixDisplay(confusion_matrix(y_test, y_preds_knn))
cmplt_knn.plot()

print(classification_report(y_preds_knn, y_test))
```

#### Support Vector Machine

```python
svm = svmc.SVC()
svm.fit(X_train, y_train)

y_pred_svm = svm.predict(X_test)

cmplt_svm = ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred_svm))
cmplt_svm.plot()

print(classification_report(y_pred_svm, y_test))
```
